```python
import pandas as pd
import numpy as np
import openpyxl
import matplotlib.pyplot as plt
raw_data=pd.read_csv("kpmi_ru_data.csv")
raw_data.head()
```

```python
result_ei=raw_data[["e","i"]]
result_ei.head()
```

```python
count=result_ei.value_counts()
count.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
x_coord=[i[0] for i in count.index]
y_coord=[j[1] for j in count.index]
z_value=count.values
```

```python
fig = plt.figure() # åˆ›å»ºä¸€ä¸ªæ–°çš„ figure
ax = fig.add_subplot(111, projection='3d') # æ·»åŠ ä¸€ä¸ªä¸‰ç»´åæ ‡ç³»
ax.scatter(x_coord, y_coord, z_value, c=z_value, cmap='viridis', marker='o')
ax.set_xlabel('Score of E')
ax.set_ylabel('Score of I')
ax.set_zlabel('Frequency')
plt.show()
```
