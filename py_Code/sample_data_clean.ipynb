### 0. Import module and pre-processing

```python
import pandas as pd
import openpyxl
import re
import string
import nltk
raw_data=pd.read_csv("Data/mbti_1.csv")
raw_data.head()
```

```python
sample=raw_data.loc[0,"posts"].split("|||")
sample
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4') # Open Multilingual Wordnet, 某些版本的 wordnet 需要
my_nltk_path="C:\\Users\\DominicMin\\AppData\\Roaming"
nltk.data.path.append(my_nltk_path)
```

### 1. Convert to lower case

```python
sample_lower=[]
#1. lower case
for i in sample:
    sample_lower.append(i.lower())
sample_lower
```

### 2. Remove URL

```python
sample_without_url=[]
for i in sample_lower:
    sample_without_url.append(
        re.sub(
            pattern=r'http\S+|www\S+|https\S+',
            repl='',
            string=i,
            flags=re.MULTILINE
        )
    )
sample_without_url
```

### 3. Remove punctuation

```python
sample_without_punct=[]
for i in sample_without_url:
    sample_without_punct.append(
        re.sub(
            pattern=r'[^\w\s]',
            repl='',
            string=i
        )
    )
sample_without_punct
```

### 4. Remove empty string and whitespace characters

```python
sample_without_whitespace=[i for i in sample_without_punct if i.strip()]
sample_without_whitespace
```

### 5. Tokenization

```python
sample_tokens=[]
for text in sample_without_whitespace:
    tokens=word_tokenize(text)
    sample_tokens.append(tokens)
sample_tokens
```

### 6. Remove stop words

```python
stop_words = set(stopwords.words('english'))
filtered_tokens=[]
for i in sample_tokens:
    temp=[]
    for j in i:
        if j not in stop_words:
            temp.append(j)
    filtered_tokens.append(temp)
filtered_tokens
```

### 7. Lemmatization

```python
# 初始化词形还原器
lemmatizer = WordNetLemmatizer()
final_data=[]
for i in filtered_tokens:
    temp=[]
    for j in i:
        temp.append(lemmatizer.lemmatize(j))
    final_data.append(temp)
final_data
```

### 8. Save result

```python
import pickle
with open('sample_cleaned.pkl', 'wb') as f:
    pickle.dump(final_data,f)
```
