```python
import pandas as pd

# 读取数据
df = pd.read_csv(r'C:\Users\微软\Desktop\archive\kpmi_ru_data.csv')

# 删除空值（虽然你说没有空值，但保险起见）
df_clean = df.dropna()

# 选取数值型列
numeric_cols = df_clean.select_dtypes(include='number').columns

# 替换异常值（使用 IQR 法则）为该列中位数
df_processed = df_clean.copy()

for col in numeric_cols:
    Q1 = df_processed[col].quantile(0.25)
    Q3 = df_processed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_value = df_processed[col].median()

    # 替换异常值
    df_processed[col] = df_processed[col].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)

# 查看结果
print(df_processed.info())
print(df_processed)
df_processed.to_csv(r'C:\Users\微软\Desktop\archive\kpmi_ru_data(Cleaned).csv', index = False)
```

```python
import numpy as np
from scipy.stats import chi2

# 1) 读入清理好的 CSV
FILE_PATH = r"C:\Users\微软\Desktop\archive\kpmi_ru_data(Cleaned).csv"   # 根据你的实际路径调整
df = pd.read_csv(FILE_PATH)

# 2) 提取四个维度的列（IE, SN, TF, JP）
mbti = df['psychotype'].astype(str).str.upper().dropna()
df_dims = pd.DataFrame({
    'IE': mbti.str[0],   # E 或 I
    'SN': mbti.str[1],   # S 或 N
    'TF': mbti.str[2],   # T 或 F
    'JP': mbti.str[3]    # J 或 P
})

# 3) 构建 2×2×2×2 列联表：得到 16 个格子的观测频数 counts[i,j,k,l]
counts = (
    df_dims.groupby(['IE','SN','TF','JP'])
    .size()
    .reindex(
        pd.MultiIndex.from_product(
            [['E','I'], ['S','N'], ['T','F'], ['J','P']],
            names=['IE','SN','TF','JP']
        ),
        fill_value=0
    )
    .values
    .reshape(2, 2, 2, 2)
)

# 4) 计算边际频数
# N_i··· 维度 IE=i 的边际总数；其余类似
N_IE = counts.sum(axis=(1,2,3))   # shape=(2,)
N_SN = counts.sum(axis=(0,2,3))   # shape=(2,)
N_TF = counts.sum(axis=(0,1,3))
N_JP = counts.sum(axis=(0,1,2))

# 总样本量 N
N = counts.sum()

# 5) 根据完全独立假设计算每个格子的期望频数 E[i,j,k,l]
E = np.zeros_like(counts, dtype=float)
for i in range(2):
    for j in range(2):
        for k in range(2):
            for l in range(2):
                # P(IE=i) = N_IE[i]/N  等等
                # E = N * P(IE=i)*P(SN=j)*P(TF=k)*P(JP=l)
                E[i,j,k,l] = (
                    (N_IE[i]/N) *
                    (N_SN[j]/N) *
                    (N_TF[k]/N) *
                    (N_JP[l]/N)
                ) * N

# 6) 计算 Pearson χ² 统计量： sum( (O - E)^2 / E ) over all 16 cells
chi2_stat = ((counts - E)**2 / E).sum()

# 7) 自由度 df = 10
df_val = 10

# 8) 计算 p 值
p_val = 1 - chi2.cdf(chi2_stat, df_val)

# 9) 输出并给出结论
print(f"Chi² 统计量 = {chi2_stat:.2f}")
print(f"自由度      = {df_val}")
print(f"p 值        = {p_val:.4g}")
if p_val < 0.05:
    print("结论：拒绝 H₀ —— 四个维度并非完全独立。")
else:
    print("结论：无法拒绝 H₀ —— 四个维度可视为独立。")
```
